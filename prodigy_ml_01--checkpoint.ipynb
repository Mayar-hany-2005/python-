```python
from sklearn.base import BaseEstimator , TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
class FeatureImputer(BaseEstimator,TransformerMixin):
    def fit(self,X,y = None):
        return self
    def transform(self,X):
        imputer = SimpleImputer(strategy = "median")
        X[["GarageYrBlt","LotFrontage","MasVnrArea","BsmtFullBath","BsmtUnfSF","BsmtFinSF1","GarageCars"]] = imputer.fit_transform(X[["GarageYrBlt","LotFrontage","MasVnrArea","BsmtFullBath","BsmtUnfSF","BsmtFinSF1","GarageCars"]])
        return X
```

```python
class FeatureEncoder(BaseEstimator, TransformerMixin):
    
    def fit(self, X, y=None):
        
        self.ordinal_encoder = OrdinalEncoder(categories=[['Fa', 'TA', 'Gd', 'Ex'], ['Fa', 'TA', 'Gd', 'Ex']])
        self.ordinal_encoder.fit(X[["ExterQual", "KitchenQual"]])
        
        self.freq_encoding = {}
        for col in ["Neighborhood", "Foundation"]:
            freq_map = X[col].value_counts(normalize=True).to_dict()
            self.freq_encoding[col] = freq_map

        self.one_hot_encoder = OneHotEncoder(drop="first", sparse=False)
        self.one_hot_encoder.fit(X[["CentralAir"]])
            
        return self
    
    def transform(self, X):
        central_air_encoded = self.one_hot_encoder.transform(X[["CentralAir"]])
        
        X = X.drop(columns=["CentralAir"])
        X[["CentralAir_Y"]] = central_air_encoded
        
        X[["ExterQual", "KitchenQual"]] = self.ordinal_encoder.transform(X[["ExterQual", "KitchenQual"]])
        
        for col in ["Neighborhood", "Foundation"]:
            X[col] = X[col].map(self.freq_encoding[col])
        
        return X
```
